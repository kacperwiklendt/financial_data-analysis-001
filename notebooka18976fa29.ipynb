```python
# Wczytanie danych
df = pd.read_csv('/kaggle/input/billionaires-statistics-dataset/Billionaires Statistics Dataset.csv')

# PodglƒÖd pierwszych 5 wierszy
df.head()
```

```python
# Informacje o danych
df.info()

# Podstawowe statystyki
df.describe()

# Sprawdzenie brak√≥w
df.isnull().sum()
```

```python
# Histogram wieku (je≈õli masz kolumnƒô 'Age')
plt.figure(figsize=(8,5))
sns.histplot(df['age'], bins=20, kde=True)
plt.title('Rozk≈Çad wieku')
plt.xlabel('Wiek')
plt.ylabel('Liczba os√≥b')
plt.show()

# S≈Çupkowy wykres p≈Çci
sns.countplot(data=df, x='gender')
plt.title('Liczba os√≥b wg p≈Çci')
plt.ylabel('Liczba os√≥b')
plt.xlabel('p≈Çeƒá')
plt.show()
```
